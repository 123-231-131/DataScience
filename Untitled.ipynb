```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('zomato_data.csv')
df.head(5)
```

let’s convert the data type of the “rate” column to float and remove the denominator.

```python
def handleRate(value):
    value = str(value).split('/')
    value = value[0]
    return float(value)
```

```python
df['rate'] = df['rate'].apply(handleRate)
df.head(15)
```

```python
df.info()
```

--> Let’s explore the listed_in (type) column.

```python
sns.countplot(x=df['listed_in(type)'])
plt.xlabel('Type of restaurant')
```

```python
grouped_data = df.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})
plt.plot(result, c = 'green', marker='o')
plt.xlabel('Type of restaurant')
plt.ylabel('Votes')
```

# 

    Conclusion: Dining restaurants are preferred by a larger number of individuals.


# Now we will determine the restaurant’s name that received the maximum votes based on a given dataframe.

```python
# .loc is a way to access specific rows and columns in a DataFrame based on labels or conditions.

max_votes = df['votes'].max()
restaurant_with_max_votes = df.loc[df['votes'] == max_votes, 'name']
print("Restaurant(s) with the maximum votes:")
restaurant_with_max_votes
```
